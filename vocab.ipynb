```python
get_ipython().run_line_magic('matplotlib', 'inline')
import pandas as pd
import numpy as np
import seaborn as sns

from datetime import datetime

from sklearn.externals import joblib 
```

```python
pkls = [
    'model/20200102T020750_smishing_0_5__3777_split.pkl',
    'model/20200102T020750_smishing_0_4__2975_split.pkl',
    'model/20200102T020750_smishing_0_3__4820_split.pkl',
    'model/20200102T020750_smishing_0_2__4123_split.pkl',
    'model/20200102T020750_smishing_0_1__4845_split.pkl',
]
```

```python
fea_dict = {}
for pkl in pkls:
    for k, v in joblib.load(pkl).items():
        if k in fea_dict.keys():
            fea_dict[k] += v
        else:
            fea_dict[k] = v
```

```python
fc_cols = []
for t in sorted(fea_dict.items(), key=(lambda x:x[1]), reverse=True)[:64]:
    fc_cols.append('_'.join(t[0].split('_')[:-1]))
```

```python
def load_csv(csv_path, c):
    df = pd.read_csv(csv_path)
    display((df['smishing'] * 10).astype(int).value_counts(sort=False))
    df.plot(kind='scatter', x='id', y='smishing', figsize=(20, 5), s=5, c=c)
    df['tag'] = csv_path[-8:-4]
    df['color'] = c
    return df
```
