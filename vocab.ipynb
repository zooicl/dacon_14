```python
fea_imp_dict = {}
for pkl in glob.glob('model/20200105T060507*split.pkl'):
    for k, v in joblib.load(pkl).items():
        if k in fea_imp_dict.keys():
            fea_imp_dict[k] += v
        else:
            fea_imp_dict[k] = v
```

```python
non_zero_cols = [k for k, v in fea_imp_dict.items() if v != 0]

zero_cols = set(fea_imp_dict.keys()) - set(non_zero_cols)

zero_dict = {}
for k, v in fea_dict.items():
    if v in zero_cols:
        zero_dict[k] = v

joblib.dump(zero_dict, '20200105T060507_zero_dict.pkl')
```

```python
def load_csv(csv_path, c):
    df = pd.read_csv(csv_path)
    display((df['smishing'] * 10).astype(int).value_counts(sort=False))
    df.plot(kind='scatter', x='id', y='smishing', figsize=(20, 5), s=5, c=c)
    df['tag'] = csv_path[-8:-4]
    df['color'] = c
    return df
```
